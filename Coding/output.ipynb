```python
# Se importan las librerías que se utilizarán en este archivo.
import pandas as pd
import json
```

```python
# Se crea una lista para posteriormente cargar allí los datos provenientes del arhcivo json, el procedimiento se realiza abriendo el archivo json  e iterando cada linea dentro del archivo para guardar linea a linea en la lista creada.
data1 = []
with open('output_steam_games.json') as f1:
    for line in f1:
        data1.append(json.loads(line))
```

```python
#creo un Data Frame a partir de la lista creada en el paso anterior.
df_output = pd.DataFrame(data1)
```

```python
#realizo un poco de EDA sobre el archivo con el fin de conocer a qué tipo de informaicón me estoy enfrentando y poder identificar columnas clave.
conteo_NaN = df_output.isna().sum().sum()
print("El número de valores NaN es:", conteo_NaN)
df_output.head(5)
```

```python
df_output.tail(5)
```

```python
# Se limpia el Data Frame de NaN y se eliminan columnas que considero irrelevantes.
df_output = df_output.dropna()
df_output = df_output.drop(['publisher', 'reviews_url','url','tags', 'app_name', 'specs', 'price', 'early_access', 'developer'], axis=1)
```

```python
#Se exporta el dataset en CSV.
df_output.head(3)
df_output.to_csv('df_outputs.csv', index=False)
```
