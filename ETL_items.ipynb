```python
# Se importan las librerías que se utilizarán en este archivo.
import pandas as pd
import json
import ast
```

```python
data3 = []
with open('australian_users_items.json', "r", encoding="utf-8") as f3:
    for line in f3.readlines():
        data3.append(ast.literal_eval(line))
df_items = pd.DataFrame(data3)
df_items.head(5)
```

```python
# Se eliminan las columnas irrelevantes
df_items = df_items.drop(['steam_id', 'user_url'], axis=1)
```

```python
df_items.head(10)
```

```python
# Se desanida la columna items para ver su información.
df_items['playtime_forever'] = df_items['items'].apply(lambda x: x[0].get('playtime_forever') if len(x) > 0 else 0)
df_items['playtime_2weeks'] = df_items['items'].apply(lambda x: x[0].get('playtime_2weeks') if len(x) > 0 else 0)
df_items['item_id'] = df_items['items'].apply(lambda x: x[0].get('item_id') if len(x) > 0 else 0)
df_items['playtime_forever'] = df_items['playtime_forever'].astype(int)
df_items['playtime_2weeks'] = df_items['playtime_2weeks'].astype(int)
df_items['item_name'] = df_items['items'].apply(lambda x: x[0].get('item_name') if len(x) > 0 else 0)
```

```python
df_items.head(10)
```

```python
df_items = df_items.drop(['user_id', 'items_count','items','playtime_forever','playtime_2weeks','id'], axis=1)
```

```python
df_items.head(10)
```

```python
df_items.to_csv('df_TitulosJuegos.csv', index=False)
```
